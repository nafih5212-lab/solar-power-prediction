```python
df.drop(['timestamp'],axis=1,inplace=True)
```

```python
df
```

```python
#change zero value as missing value
df.loc[df.ghi_W_m2 == 0,'ghi_W_m2']=np.nan
df.loc[df.pv_power_kW == 0,'pv_power_kW']=np.nan
```

```python
df['ghi_W_m2'].unique()
```

```python
df['pv_power_kW'].unique()
```

```python
df.isna().sum()
```

```python
df['ghi_W_m2']=df['ghi_W_m2'].fillna(df['ghi_W_m2'].mean())
```

```python
df['pv_power_kW']=df['pv_power_kW'].fillna(df['pv_power_kW'].mean())
```

```python
target = None   

for col in df.columns:


    if target and col == target:
        continue

    plt.figure()
    plt.plot(df[col])    
    plt.title(f"{col} Plot")
    plt.xlabel("Index")
    plt.ylabel(col)
    plt.show()
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
df3=pd.DataFrame({"actual value":y_test,"predicted value":y_pred,"difference":y_test-y_pred})
df3
```

```python
from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(y_test,y_pred)
MAE
```

```python
from sklearn.metrics import mean_absolute_percentage_error
MAPE=mean_absolute_percentage_error(y_test,y_pred)
MAPE
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_test,y_pred)
MSE
```

```python
from sklearn.metrics import root_mean_squared_error
RMSE=root_mean_squared_error(y_test,y_pred)
RMSE
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```
